```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Step 1: Send an HTTP request to the webpage using the requests library.
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html";
data = requests.get(url);
print(data);
print(type(data));# we have used .text method for extracting the HTML content as a string in order to make it readable.
html = data.text;
print(type(html));
#print(html);
```

```python
# Step 2: Parse the HTML content of the webpage using BeautifulSoup.
soup = BeautifulSoup(html, "html5lib"); # need html(not data)
#print(soup);
```

```python
# Step 3: Identify the HTML tags that contain the data you want to extract.
# Scrapping the content of the HTML webpage and convert the table into a dataframe
netflix_data = pd.DataFrame(columns = ["Date", "Open", "High", "Low", "Close", "Volume"]);
print(netflix_data);
```

```python
# Step 4: Use BeautifulSoup methods to extract the data from the HTML tags.

# First we isolate the body of the table which contains all the information
# Then we loop through each row and find all the column values for each row
for row in soup.find("tbody").find_all("tr"): # <tbody> tag: This is the main content of the table, which is defined using the tag. It contains one or more rows of elements.
  col = row.find_all("td");
  date = col[0].text; # convert to type string
  open = col[1].text;
  high = col[2].text;
  low = col[3].text;
  close = col[4].text;
  adj_close = col[5].text;
  volume = col[6].text;
  # Finally we append the data of each row to the table
netflix_data = netflix_data.append({"Date":date, "Open":open, "High":high, "Low":low, "Close":close, "Adj Close":adj_close, "Volume":volume}, ignore_index=True);
print(netflix_data);
```

#### Using Panda Library

```python
# ******** pd.read_html(url) is a function provided by the pandas library in Python that is used to extract tables from HTML web pages.
# It takes in a URL as input and returns a list of all the tables found on the webpage.
read_html_pandas_data = pd.read_html(url);
netflix_dataframe = read_html_pandas_data[0]; # There is only one table on the page, we just take the first table in the list returned

netflix_dataframe.head()
```
